```python
import fastf1
import pandas as pd

# Enable cache for fastf1
fastf1.Cache.enable_cache("/Users/ainefernandez/documents/itam/8SEMESTRE/F1 Python")

def count_safety_car_periods(laps):
    """
    Count the number of safety car periods during a race.

    Args:
        laps (DataFrame): DataFrame containing lap data.

    Returns:
        int: Number of safety car periods.
    """
    # Initialize variables to track safety car periods
    safety_car_occurred = False
    safety_car_count = 0
    
    # Iterate through each lap
    for lap in laps.itertuples():
        track_status = lap.TrackStatus
        
        # Check if track status indicates safety car
        if isinstance(track_status, str) and "4" in track_status:
            if not safety_car_occurred:
                # Start of a new safety car period
                safety_car_occurred = True
        else:
            if safety_car_occurred:
                # End of the current safety car period
                safety_car_occurred = False
                safety_car_count += 1
    
    return safety_car_count

# Define the year and Grand Prix to test
year = 2023
gp = "Australia"

try:
    # Load the session
    session = fastf1.get_session(gp, year, 'R')
    session.load_laps()
    
    # Extract lap data
    laps_data = session.laps
    
    # Count safety car periods
    safety_car_count = count_safety_car_periods(laps_data)
    print("Number of safety car periods in", gp, year, ":", safety_car_count)

except Exception as e:
    # Handle exceptions (e.g., if the session data is not available)
    print(f"Could not load data for {gp} {year}: {e}")
```
