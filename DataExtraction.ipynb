## Lap Data

Obtain lap-by-lap data from every Grand Prix race (excluding sprint races) from 2019 to 2023. 

Each record includes the following variables: Driver, Team, LapNumber, LapTime, Stint, Compound, TyreLife, and Position. 

The dataframe only includes quick laps, filtered according to the Formula 1 107% rule. This ensures that only competitive lap times are retained in the final dataset.

```python
#List of races by year
gps_by_year = {
    2019: ["Bahrain", "China", "Australia", "Japan", "Monaco", "Spain", "Canada", 
           "Austria", "Great Britain", "Hungary", "Belgium", "Azerbaijan", 
           "Singapore", "Austin", "Mexico", "Brazil", "Abu Dhabi", "Monza"],
    2020: ["Bahrain", "Monaco", "Spain", "Canada", "Austria", "Great Britain", 
           "Hungary", "Belgium", "Abu Dhabi", "Monza"],
    2021: ["Bahrain", "Monaco", "Spain", "Imola", "Austria", 
           "Great Britain", "Hungary", "Netherlands",
           "Belgium", "Azerbaijan",  "Austin", "Mexico", 
           "Brazil", "Abu Dhabi", "Monza", "Qatar","Saudi Arabia"],
    
    2022: ["Bahrain", "Australia", "Japan", "Monaco", "Spain", 
           "Canada", "Imola", "Austria", "Great Britain", "Hungary","Netherlands", 
           "Belgium", "Azerbaijan", "Singapore", "Austin", "Mexico", 
           "Brazil", "Abu Dhabi", "Monza", "Miami","Saudi Arabia"],
    
    2023: ["Bahrain", "Saudi Arabia", "Australia", "Japan",  
           "Miami", "Monaco", "Spain", "Canada", "Austria", 
           "Great Britain", "Hungary", "Belgium", "Netherlands", 
           "Azerbaijan", "Singapore", "Austin", "Mexico", "Brazil", 
           "Qatar", "Abu Dhabi", "Monza"]
}

drivers=["SAR","ALB","MAG","TSU","RIC"
         "ZHO","BOT","PIA","HUL","NOR",
         "GAS","OCO","STR","LEC","SAI",
         "RUS","HAM","ALO","PER","VER"]

# Create an empty DataFrame to store all lap data
all_lapdata = pd.DataFrame(columns=['Driver', 'Team','LapNumber', 'LapTime', 'Stint', 'Compound', 'TyreLife', 'Position', 'Year', 'GP'])

for year, gps in gps_by_year.items():
    for gp in gps:
        try:
            #Load session for each GP race
            session = fastf1.get_session(year, gp, 'R')
            session.load()

            # Retrieve lap data
            for driver in drivers:
                # Retrieve lap data, only quicklaps
                laps = session.laps.pick_driver(driver).pick_quicklaps()
                #Convert LapTime to seconds 
                laps['LapTime'] = laps['LapTime'].dt.total_seconds()
                laps = laps[['Driver', 'Team','LapNumber', 'LapTime', 'Stint', 'Compound', 'TyreLife', 'Position']]
                laps['Year'] = year
                laps['GP'] = gp

            # Append to all_lapdata
            all_lapdata = pd.concat([all_lapdata, laps], ignore_index=True)

        except Exception as e:
            print(f"No data found for {gp} {year}: {e}")
```

## LapTimePerKM and RacePercentage

Two additional variables are added to the dataset:

- **RacePercentage**: Represents the percentage of the race completed at the time each lap was set.

```python
#Filter the dataset 
dry=["SOFT","MEDIUM","HARD"]
dryquicklaps=fulldf[fulldf['Compound'].isin(dry)]

#Add variables 
dryquicklaps["LapTimePerKM"]=dryquicklaps["LapTime"]/dryquicklaps["Length"]
dryquicklaps["RacePercentage"]=dryquicklaps["LapNumber"]/dryquicklaps["Laps"]
```

## Strategy

Obtain the tyre strategy used by each driver in every Grand Prix in the dataset. 

The strategy is defined as the sequence of dry compounds (SOFT, MEDIUM, HARD) used across their stints during the race.

```python
strategy_data = []

# Iterate over each year and GP
for year, gps in gps_by_year.items():
    for gp in gps:
        try:
            # Load the race session
            session = ff1.get_session(year, gp, 'R')
            session.load()
            
            # Extract laps data
            laps = session.laps
            
            # Calculate the number of pit stops by checking PitOutTime > 0 and not the first lap
            laps['PitOutTime'] = laps['PitOutTime'].dt.total_seconds()
            pit_stops = laps[(laps['PitOutTime'] > 0) & (laps['LapNumber'] > 1)]
            
            # Extract stint and compound information for each driver
            drivers = session.drivers
            drivers = [session.get_driver(driver)["Abbreviation"] for driver in drivers]
            
            for driver in drivers:
                driver_laps = laps[laps['Driver'] == driver]
                
                # Group by stints and extract compound strategy
                stints_driver = driver_laps.groupby('Stint').first().reset_index()
                strategy = '-'.join(stints_driver['Compound'].values)  # For example: MEDIUM-HARD-HARD
                
                # Count the number of pit stops for the driver
                num_pitstops = pit_stops[pit_stops['Driver'] == driver].shape[0]
                
                # Store the data
                strategy_data.append({
                    'Year': year,
                    'GP': gp,
                    'Driver': driver,
                    'Strategy': strategy,
                    'PitStops': num_pitstops
                })
        
        except Exception as e:
            print(f"Error processing {gp} {year}: {e}")

# Create a DataFrame with all the data
strategy_df = pd.DataFrame(strategy_data)

stint_data = []


for year, gps in gps_by_year.items():
    for gp in gps:
        try:
            # Load the race session
            session = ff1.get_session(year, gp, 'R')
            session.load()
            
            # Extract laps data
            laps = session.laps
            
            # Extract stint and compound information for each driver
            drivers = session.drivers
            drivers = [session.get_driver(driver)["Abbreviation"] for driver in drivers]
            
            for driver in drivers:
                driver_laps = laps[laps['Driver'] == driver]
                
                # Calculate the stint length for each stint
                stint_lengths = driver_laps.groupby(['Stint', 'Compound']).size().reset_index(name='StintLength')
                
                # Add StintNumber and driver info
                stint_lengths['StintNumber'] = stint_lengths['Stint']
                stint_lengths['Year'] = year
                stint_lengths['GP'] = gp
                stint_lengths['Driver'] = driver
                
                # Store stint data
                stint_data.append(stint_lengths)
        
        except Exception as e:
            print(f"Error processing {gp} {year}: {e}")

# Create a DataFrame for stint data
stint_df = pd.concat(stint_data)

# Merge the stint data with your existing strategy_df
final_df = pd.merge(strategy_df, stint_df, on=['Year', 'GP', 'Driver'], how='left')

final_df.to_csv("TesisStrategyfull.csv")
```

## Pitstop Time

Pitstop time is defined as the duration from the moment a driver enters the pit lane until the car exits the pit lane.

### Circuit and Team Mapping

As circuit names are stored differently in the Ergast API compared to the lap dataset, a mapping is required to standardize them.  

Additionally, a driver-to-team mapping is applied to assign the correct team to each pitstop record.


```python
pitstopsall = pd.DataFrame(columns=["GP", "Circuit", "PitstopT", "Driver", "Year"])
years = [2019, 2020, 2021, 2022, 2023]
nraces = [21, 17, 22, 22, 22]

for year, races in zip(years, nraces): 
    for i in range(1, races + 1):
        try:
            pitstops = ergast.get_pit_stops(year, i, result_type="raw", limit=200, auto_cast=True)[0]
            circuit = pitstops['Circuit']["circuitId"]
            for j in range(0, len(pitstops["PitStops"]) - 1):
                driver = pitstops["PitStops"][j]['driverId']
                duration = pitstops["PitStops"][j]['duration']
                p = {"GP": i, "Circuit": circuit, "PitstopT": duration, "Driver": driver, "Year": year}
                pitstopsall = pitstopsall.append(p, ignore_index=True)
        except Exception as e: 
            print(f"No data found for {i} {year}: {e}")

```

## Outlaps

The outlap is the first lap a driver completes after a pitstop.  

It is identified in the data by a positive value in the PitOutTime variable.


```python
# Create an empty DataFrame
outlaps = pd.DataFrame(columns=['Driver', 'Team', 'LapNumber', 'LapTime', 'Stint', 'Compound', 'TyreLife', 'Position', 'Year', 'GP', 'IsAccurate'])

for year, gps in gps_by_year.items():
    for gp in gps:
        try:
            session = fastf1.get_session(year, gp, 'R')
            session.load()
            drivers = pd.unique(session.laps['Driver'])
            for driver in drivers:
                # Retrieve lap data
                laps = session.laps.pick_driver(driver)
                laps['LapTime'] = laps['LapTime'].dt.total_seconds()
                laps['PitInTime'] = laps['PitInTime'].dt.total_seconds()
                laps['PitOutTime'] = laps['PitOutTime'].dt.total_seconds()
                laps = laps[laps['PitOutTime'] > 0]
                laps = laps[laps['LapNumber'] != 1]
                
                # Select desired columns
                laps = laps[['Driver', 'Team', 'LapNumber', 'LapTime', 'Stint', 'Compound', 'TyreLife', 'Position', 'IsAccurate']]
                # Add year and GP
                laps['Year'] = year
                laps['GP'] = gp
                # Append to outlaps
                outlaps = outlaps.append(laps, ignore_index=True)

        except Exception as e:
            print(f"No data found for {gp} {year}: {e}")


circuit_info = pd.read_csv("TesisCircuitInfo.csv")

#Circuit mapping
circuit_mapping = {
    '70th': 'Great Britain',
    'Sakhir': 'Bahrain',
    'Styria': 'Austria'
}

#Team mapping
team_mapping = {
    'Toro Rosso': 'AlphaTauri',
    'Renault': 'Alpine',
    'Alfa Romeo Racing': 'Alfa Romeo',
    'Racing Point': 'Aston Martin'
}

outlaps['GP'] = outlaps['GP'].replace(circuit_mapping)
outlaps['Team'] = outlaps['Team'].replace(team_mapping)

# Filter out rows where 'Team' is NaN
outlaps = outlaps[outlaps['Team'].notna()]

# Merge outlaps with circuit info
outlaps = pd.merge(outlaps, circuit_info, on='GP', how='left')

# Filter outlaps for dry compound types
outlaps = outlaps[outlaps['Compound'].isin(dry)]

# Calculate LapTimePerKM
outlaps["LapTimePerKM"] = outlaps["LapTime"] / outlaps["Length"]

outlaps.to_csv("TesisOutlaps.csv", index=False)
```

## Inlaps 

The inlap is the  lap when a driver comes into the pits.

It is identified in the data by a positive value in the PitInTime variable.

```python
inlaps = pd.DataFrame(columns=['Driver', 'Team', 'LapNumber', 'LapTime', 'Stint', 'Compound', 'TyreLife', 'Position', 'Year', 'GP','IsAccurate'])

for year in gps_by_year:
    for gp in gps_by_year[year]:
        try:
            session = fastf1.get_session(year, gp, 'R')
            session.load()
            drivers = pd.unique(session.laps['Driver'])
            for driver in drivers:
                # Retrieve lap data for the specific driver
                laps = session.laps.pick_driver(driver)
                laps['LapTime'] = laps['LapTime'].dt.total_seconds()
                laps['PitInTime'] = laps['PitInTime'].dt.total_seconds()
                laps['PitOutTime'] = laps['PitOutTime'].dt.total_seconds()
                
                # Filter laps for valid in-laps
                laps = laps[laps['PitInTime'] > 0]
                laps = laps[laps['LapNumber'] != 1]
                
                # Select desired columns
                laps = laps[['Driver', 'Team', 'LapNumber', 'LapTime', 'Stint', 'Compound', 'TyreLife', 'Position', 'IsAccurate']]
                
                # Add year and GP columns
                laps['Year'] = year
                laps['GP'] = gp
                
                # Append to inlaps DataFrame
                inlaps = inlaps.append(laps, ignore_index=True)
        
        except Exception as e:
            print(f"Could not process {gp} {year}: {str(e)}")

inlaps['GP'] = inlaps['GP'].replace(circuit_mapping)
inlaps['Team'] = inlaps['Team'].replace(team_mapping)

# Remove rows where 'Team' column is NaN
inlaps = inlaps.dropna(subset=['Team'])

# Merge inlaps with circuit info
inlaps = pd.merge(inlaps, circuit_info, on='GP', how='left')

# Filter inlaps for dry compound types
inlaps = inlaps[inlaps['Compound'].isin(dry)]

# Calculate LapTimePerKM
inlaps["LapTimePerKM"] = inlaps["LapTime"] / inlaps["Length"]


inlaps.to_csv("TesisInlaps.csv", index=False)

```
